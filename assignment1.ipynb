```python
# Task 2: Data Exploration
# 1. Choose 1 column with nominal values, 1 column with ordinal Values, and 1 column with numerical values


#Plot pie chart of 'Body-style' column
Automobile['Body-style'].value_counts()\
                        .plot(kind='pie', autopct='%1.1f%%', radius=1.2
                              ,textprops = {"fontsize":13})
plt.title("Pie chart of Body-Style", bbox={'facecolor':'0.9', 'pad':5}, pad=30)
plt.show() 
```

```python
#Plot bar plot of the 'Symboling' column
Symbolplot=Automobile['Symboling'].value_counts().sort_index()
plt.bar(Symbolplot.index, Symbolplot.values, color ='orange',
        width = 0.5)
plt.xlabel("Insurance risk rating")
plt.ylabel("Frequency")
plt.title("Insurance risk rating bar plot")
plt.show()
```

```python
#Plot histogram plot of the 'Price'
Automobile['Price'].plot(kind='hist', bins = 10)
plt.title('Price distribution')
plt.xlabel('Price')
plt.grid()
plt.show()
```

```python
# Task 2: Data Exploration
# 2. Explore the relationships between columns; at least 3 visualisations with plausible hypothesis

#Code goes after this line
#Import seaborn into memory and name it sns
import seaborn as sns
#set the plot size
plt.figure(figsize=(16,6))
#plot a boxplot of "Engine-size" column group by "Num-of-cylinder" column and arrange the order of the box
sns.boxplot(y=Automobile['Engine-size'], x=Automobile['Num-of-cylinders'],
            order=["two", "three", "four", "five", "six", "eight", "twelve"])
plt.title('Boxplot of engine-size group by number of cylinder', bbox={'facecolor':'0.9', 'pad':5}, pad=15)
plt.show()
```

```python
#Plot a scatter plot between the "Curb-weight" column and the "Highway-mpg"
sns.scatterplot(data=Automobile, x="Curb-weight", y="Highway-mpg")
#Add title box
plt.title('Scatter plot of the curb weight and the highway Miles Per Gallon', bbox={'facecolor':'0.9', 'pad':5}, pad=15)
plt.show()
```

```python
#https://seaborn.pydata.org/generated/seaborn.scatterplot.html
sns.scatterplot(data=Automobile, x="Horsepower", y="City-mpg")
plt.title('Scatter plot of the horsepower and the city miles per gallon of the car', bbox={'facecolor':'0.9', 'pad':5}, pad=15)
plt.show()
```
